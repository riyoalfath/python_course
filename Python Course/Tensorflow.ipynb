```python
import tensorflow as tf
```

```python
vector = tf.constant([1, 2, 3, 4, 5])
scalar = tf.constant(12)
matrix = tf.constant([[1, 2, 3], [4, 5, 6]])
tensor = tf.constant([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
```

```python
print("Shape:", tensor.shape)
print("Rank:", tf.rank(tensor))
print("Size:", tf.size(tensor))
```

```python
tensor1 = tf.random.uniform([5, 300], minval=0, maxval=1)
tensor2 = tf.random.uniform([5, 300], minval=0, maxval=1)

print(tensor1)
print(tensor2)
```

```python
matmulResult = tf.matmul(tensor1, tf.transpose(tensor2))

print(matmulResult.shape)
```

```python
dotProduct = tf.tensordot(tensor1, tensor2, axes=[[1], [1]])

print(dotProduct.shape)
```

```python
imageTensor = tf.random.uniform([224, 224, 3], minval=0, maxval=1)

print(imageTensor)
```

```python
minVal = tf.reduce_min(imageTensor)
maxVal = tf.reduce_max(imageTensor)

print(minVal)
print(maxVal)
```

```python
tensor3 = tf.random.uniform([1, 224, 224, 3], minval=0, maxval=1)
squeezedTensor = tf.squeeze(tensor3)

print(tensor3)
print(squeezedTensor)
```

```python
tensor10 = tf.constant([10, 40, 5, 3, 15, 7, 25, 8, 81, 12])
maxIndex = tf.argmax(tensor10)

print(tensor10)
print(maxIndex)
```

```python
oneHotEncoded = tf.one_hot(tensor10, depth=26)

print(oneHotEncoded)
```
