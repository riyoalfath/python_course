```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("cars1.csv", header=None)

# Naming column in dataset
df.columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']

# Tampilkan contoh data
print(df.head())
```

```python
# Take random 50 sample
df_sample = df # I using my own sample //this code for random: .sample(n=50, random_state=42)
print(df_sample)
```

```python
# Dif train data (40 data) and test data (10 data)
train_data = df_sample.iloc[:40, :]
test_data = df_sample.iloc[40:, :]

X_train = train_data.drop('class', axis=1)
y_train = train_data['class']

X_test = test_data.drop('class', axis=1)
y_test = test_data['class']

print(X_train, y_train)
print(X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

# Category into numerical
le = LabelEncoder()
for col in X_train.columns:
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
    print(X_test, X_train)

# Initiation model Naive Bayes
# Can use GaussianNB, CategoricalNB, ComplementNB, MultinomialNB --not sure what different of them
model = GaussianNB()

# Train model
model.fit(X_train, y_train)

# Predict test model
y_pred = model.predict(X_test)

```

```python
# Evaluate model
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=0)

print("Akurasi:", accuracy)
print("Laporan Klasifikasi:\n", report)
```
